```python
import matplotlib.pyplot as plt
import numpy as np

no_days = 30

days = np.linspace(0, no_days, 1000)
sine_wave = np.sin((days / no_days) * 2 * np.pi)
cosine_wave = np.cos((days / no_days) * 2 * np.pi)

plt.figure(figsize=(12, 6))
plt.plot(days + 1, sine_wave, label='Sine Wave', color='blue')
plt.plot(days + 1, cosine_wave, label='Cosine Wave', color='green')

plt.scatter([1, 16], [np.sin(0), np.sin(np.pi)], color='orange', marker='o', s=50, zorder=5)
plt.scatter([1, 16], [np.cos(0), np.cos(np.pi)], color='red', marker='o', s=50, zorder=5)

plt.axvline(x=1, linestyle='--', linewidth=1)
plt.axvline(x=no_days + 1, linestyle='--', linewidth=1)

plt.xticks(np.arange(1, no_days+2, 1))

plt.xlabel('Days')
plt.ylabel('Value')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()

# list(range(31))
```
